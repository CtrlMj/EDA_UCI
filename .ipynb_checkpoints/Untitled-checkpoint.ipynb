##### I am interested to see if I can estimate the final score itself (not what class it falls into). Therefore this is a regression problem and when the final output is a number, I don't think any of these features can be ignored.

##### transforming categorical features to numeric values:

```python
le = LabelEncoder()
df['school'] = le.fit_transform(df['school'])
df['sex'] = le.fit_transform(df['sex'])
df['address'] = le.fit_transform(df['address'])
df['famsize'] = le.fit_transform(df['famsize'])
df['Pstatus'] = le.fit_transform(df['Pstatus'])
df['famsup'] = le.fit_transform(df['famsup'])
df['schoolsup'] = le.fit_transform(df['schoolsup'])
df['paid'] = le.fit_transform(df['paid'])
df['activities'] = le.fit_transform(df['activities'])
df['nursery'] = le.fit_transform(df['nursery'])
df['higher'] = le.fit_transform(df['higher'])
df['internet'] = le.fit_transform(df['internet'])
df['romantic'] = le.fit_transform(df['romantic'])
```

```python
df = student_mat.drop(['G1', 'G2', 'G3'], axis=1)
targets = student_mat[['G1', 'G2', 'G3']]
df.head()
```

```python
le = LabelEncoder()
df['school'] = le.fit_transform(df['school'])
df['sex'] = le.fit_transform(df['sex'])
df['address'] = le.fit_transform(df['address'])
df['famsize'] = le.fit_transform(df['famsize'])
df['Pstatus'] = le.fit_transform(df['Pstatus'])
df['famsup'] = le.fit_transform(df['famsup'])
df['schoolsup'] = le.fit_transform(df['schoolsup'])
df['paid'] = le.fit_transform(df['paid'])
df['activities'] = le.fit_transform(df['activities'])
df['nursery'] = le.fit_transform(df['nursery'])
df['higher'] = le.fit_transform(df['higher'])
df['internet'] = le.fit_transform(df['internet'])
df['romantic'] = le.fit_transform(df['romantic'])
df = pd.get_dummies(df)
df.info()
```

```python
tr = DecisionTreeRegressor()
cross_val_score(tr, X=df, y=targets, cv=6)
```

```python
df = pd.get_dummies(df, columns=['Mjob', 'Fjob', 'reason', 'guardian'])
mlr = MultiOutputRegressor(LinearRegression())
cross_val_score(mlr, X=df, y=targets, cv=6)
```

```python
cross_val_score(tr, X=df, y=targets, cv=6)
```
